```python
import yfinance as yf 
import datetime 
import pandas as pd 
import numpy as np 
pd.core.common.is_list_like = pd.api.types.is_list_like #
import pandas_datareader as pdr 
yf.pdr_override()
```

```python
def get_stocks():
    f = open('Stocks.txt')
    stocks = []
    for line in f:
        stocks.append(line[0:len(line)-1])
    f.close()
    return stocks
```

```python
stocks = get_stocks()

start = datetime.datetime(2017, 1, 1) 
end = datetime.datetime(2018, 1, 1) 
data = yf.download(stocks, start=start, end=end) 
print(data)
```
